```python
import time
start_time = time.time()

def ubahhuruf(grid):
    for i in range(len(grid)):
        for n in range(len(grid[i])):
            if grid[i][n] == 1:
                grid[i][n] = 'a'
            elif grid[i][n] == 2:
                grid[i][n] = 'b'
            elif grid[i][n] == 3:
                grid[i][n] = 'c'
            elif grid[i][n] == 4:
                grid[i][n] = 'd'
            elif grid[i][n] == 5:
                grid[i][n] = 'e'
            elif grid[i][n] == 6:
                grid[i][n] = 'f'
            elif grid[i][n] == 7:
                grid[i][n] = 'g'
            elif grid[i][n] == 8:
                grid[i][n] = 'h'
            elif grid[i][n] == 9:
                grid[i][n] = 'i'
            else :
                grid[i][n] = ''
    return grid


def ubahangka(grid):
    for i in range(len(grid)):
        for n in range(len(grid[i])):
            if grid[i][n] == 'a':
                grid[i][n] = 1
            elif grid[i][n] == 'b':
                grid[i][n] = 2
            elif grid[i][n] == 'c':
                grid[i][n] = 3
            elif grid[i][n] == 'd':
                grid[i][n] = 4
            elif grid[i][n] == 'e':
                grid[i][n] = 5
            elif grid[i][n] == 'f':
                grid[i][n] = 6
            elif grid[i][n] == 'g':
                grid[i][n] = 7
            elif grid[i][n] == 'h':
                grid[i][n] = 8
            elif grid[i][n] == 'i':
                grid[i][n] = 9
            else :
                grid[i][n] = 0
    return grid


def str_to_puzzle(s):
    puzzleSolution = []
    for i in range(len(s)):  
        if i % 9 == 0:
            temp = []
            for j in s[i:i+9]:
                temp.append(int(j))
            puzzleSolution.append(temp)
    return puzzleSolution
    
def same_row(i,j):
    if i//9 == j//9:
        return True
    return False

def same_col(i,j):
    if i%9 == j%9:
        return True
    return False

def same_block(i,j):
    if ((i//9)//3 == (j//9)//3) & ((i%9)//3 == (j%9)//3):
        return True
    return False

def sudoku_brute_force(s):
    #1
    i = s.find('0')

    #2
    cannotuse = {s[j] for j in range(len(s)) if same_row(i, j) | same_col(i, j) | same_block(i, j)}
    every_possible_values = {str(i) for i in range(10)} - cannotuse

    #3
    solusi = []
    for val in every_possible_values:
        s = s[0:i] + val + s[i+1: ]
        sudoku_brute_force(s)
        if s.find('0') == -1:
            str_to_puzzle(s)
            solusi = str_to_puzzle(s)
            ubahhuruf(solusi)
            print(*solusi, sep='\n')

#masukan Grid dibawah sini
#Grid 1
puzzleToSolve =  [[5, 3, 0, 0, 7, 0, 0, 0, 0],
                  [6, 0, 0, 1, 9, 5, 0, 0, 0],
                  [0, 9, 8, 0, 0, 0, 0, 6, 0],
                  [8, 0, 0, 0, 6, 0, 0, 0, 3],
                  [4, 0, 0, 8, 0, 3, 0, 0, 1],
                  [7, 0, 0, 0, 2, 0, 0, 0, 6],
                  [0, 6, 0, 0, 0, 0, 2, 8, 0],
                  [0, 0, 0, 4, 1, 9, 0, 0, 5],
                  [0, 0, 0, 0, 8, 0, 0, 7, 9]]

#Grid 2
#puzzleToSolve =  [[3, 0, 0, 8, 0, 1, 0, 0, 2],
                  #[2, 0, 1, 0, 3, 0, 6, 0, 4],
                  #[0, 0, 0, 2, 0, 4, 0, 0, 0],
                  #[8, 0, 9, 0, 0, 0, 1, 0, 6],
                  #[0, 6, 0, 0, 0, 0, 0, 5, 0],
                  #[7, 0, 2, 0, 0, 0, 4, 0, 9],
                  #[0, 0, 0, 5, 0, 9, 0, 0, 0],
                  #[9, 0, 4, 0, 8, 0, 7, 0, 5],
                  #[6, 0, 0, 1, 0, 7, 0, 0, 3]]

#Grid 3
#puzzleToSolve =  [[0, 6, 0, 1, 0, 4, 0, 5, 0],
#                  [0, 0, 8, 3, 0, 5, 6, 0, 0],
#                  [2, 0, 0, 0, 0, 0, 0, 0, 1],
#                  [8, 0, 0, 4, 0, 7, 0, 0, 6],
#                  [0, 0, 6, 0, 0, 0, 3, 0, 0],
#                  [7, 0, 0, 9, 0, 1, 0, 0, 4],
#                  [5, 0, 0, 0, 0, 0, 0, 0, 2],
#                  [0, 0, 7, 2, 0, 6, 9, 0, 0],
#                  [0, 4, 0, 5, 0, 8, 0, 7, 0]]

s = ''.join(map(str,[''.join(map(str, i)) for i in puzzleToSolve]))

ubahhuruf(puzzleToSolve)
print("Sudoku Problem")
print(*puzzleToSolve, sep='\n')
print("\nSudoku Solution")
x = sudoku_brute_force(s)
print("Process Grid 1 finished --- %s seconds ---" % (time.time() - start_time))
```
